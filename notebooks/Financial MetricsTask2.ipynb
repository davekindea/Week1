```python
# Calculate rolling mean and standard deviation
stock_data_AAPL['Rolling_Mean'] = stock_data_AAPL['Close'].rolling(window=20).mean()
stock_data_AAPL['Rolling_Std'] = stock_data_AAPL['Close'].rolling(window=20).std()

# Bollinger Bands
stock_data_AAPL['Upper_Band'] = stock_data_AAPL['Rolling_Mean'] + (2 * stock_data_AAPL['Rolling_Std'])
stock_data_AAPL['Lower_Band'] = stock_data_AAPL['Rolling_Mean'] - (2 * stock_data_AAPL['Rolling_Std'])
```
